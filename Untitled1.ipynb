```python
import json
import re
from textblob import TextBlob

def clean_text(text):
   # Remove special characters and digits
   text = re.sub(r'[^a-zA-Z\s]', '', text)

   # Convert to lowercase
   text = text.lower()

   # Remove extra whitespace
   text = ' '.join(text.split())

   return text

def process_json_objects(input_file, output_file):
   all_cleaned_texts = []

   try:
       with open(input_file, 'r') as file:
           content = file.read()

       # Find all JSON objects in the file
       json_objects = re.findall(r'\{[^{}]*\}', content)

       for i, json_str in enumerate(json_objects):
           try:
               data = json.loads(json_str)
               review_text = data.get('review_text', '')

               if not review_text:
                   print(f"Warning: 'review_text' field is empty or missing in object {i+1}.")
                   continue

               # Clean and correct the text
               cleaned_text = clean_text(review_text)
               corrected_text = cleaned_text

               all_cleaned_texts.append(corrected_text)

               print(f"Processed object {i+1}")

           except json.JSONDecodeError as e:
               print(f"Error decoding JSON object {i+1}: {str(e)}")

       # Save all cleaned texts to a single JSON file
       with open(output_file, 'w') as outfile:
           json.dump({'cleaned_texts': all_cleaned_texts}, outfile, indent=2)

       print(f"All cleaned texts have been saved to {output_file}")

   except FileNotFoundError:
       print(f"Error: The file '{input_file}' was not found.")
   except Exception as e:
       print(f"An unexpected error occurred: {e}")

# Example usage
input_file = 'https://drive.google.com/file/d/1VbOYtm4Zh7qjjCbr2gp38O1QcFof91Zd/view?usp=drive_link'
#input_file = '/content/drive/MyDrive/A.json'
#input_file = 'A.json'
output_file = 'all_cleaned_texts.json'

process_json_objects(input_file, output_file)

print(f"Processing complete. Check '{output_file}' for all cleaned texts.")


```

```python
import json
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords

# Download stopwords if not already downloaded
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def load_cleaned_texts(filename):
    with open('all_cleaned_texts.json', 'r') as file:
        data = json.load(file)
        return data['cleaned_texts']

def tokenize_texts(texts):
    tokens = []
    for text in texts:
        words = text.split()  # Split each text into individual words
        for word in words:
            if word not in stop_words:  # Check if word is not a stopword
                tokens.append(word)  # Add to tokens if not a stopword
    return tokens

def plot_word_frequencies(frequencies):
    # Create a figure with subplots
    fig, axs = plt.subplots(1, 2, figsize=(15, 5))

    # Word Cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(frequencies)
    axs[0].imshow(wordcloud, interpolation='bilinear')
    axs[0].axis('off')
    axs[0].set_title('Word Cloud')

    # Bar Chart
    words, counts = zip(*frequencies.most_common(20))
    axs[1].bar(words, counts)
    axs[1].set_xlabel('Words')
    axs[1].set_ylabel('Frequency')
    axs[1].set_xticks(range(len(words)))
    axs[1].set_xticklabels(words, rotation=45)
    axs[1].set_title('Top 20 Most Frequent Words')

    # Adjust layout and display both plots
    plt.tight_layout()
    plt.show()

# Main Execution
cleaned_texts = load_cleaned_texts('all_cleaned_texts.json')
tokens = tokenize_texts(cleaned_texts)
word_frequencies = Counter(tokens)
plot_word_frequencies(word_frequencies)
```

```python
from textblob import TextBlob
import json

def analyze_sentiment(input_file, output_file):
    try:
        with open(input_file, 'r') as file:
            data = json.load(file)

        cleaned_texts = data.get('cleaned_texts', [])

        sentiment_scores = []
        for i, text in enumerate(cleaned_texts):
            blob = TextBlob(text)
            sentiment_score = blob.sentiment.polarity  # TextBlob returns polarity (-1 to 1)
            sentiment_scores.append({
                'review': text,
                'sentiment_score': sentiment_score
            })
            print(f"Review {i+1}: Sentiment Score = {sentiment_score}")

        with open(output_file, 'w') as outfile:
            json.dump(sentiment_scores, outfile, indent=2)

        print(f"Sentiment analysis complete. Results saved to {output_file}")

    except FileNotFoundError:
        print(f"Error: The file '{input_file}' was not found.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Example usage
input_file = 'all_cleaned_texts.json'
output_file = 'sentiment_analysis_results.json'

analyze_sentiment(input_file, output_file)
```
